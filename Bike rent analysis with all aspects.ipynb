```python
# methods for creating the count plot based on hour for a given year
def plot_by_hour(data, year=None, agg='sum'):
    dd=data
    if year: dd = dd[ dd.datetime.dt.year ==year]
    dd.loc[:, ('hour')] = dd.datetime.dt.hour      # extracting the hour data if the year in the data is equal to the year passed
    by_hour = dd.groupby(['hour', 'workingday'])['count'].agg(agg).unstack()   #Group by hour and working day
    return by_hour.plot(kind = 'bar', ylim= (0,80000), figsize = (15,5), width =0.9, title = 'year = {0}'.format(year))

plot_by_hour(train, year=2011)       #plotting the count plot based on hour for 2011
plot_by_hour(train, year=2012)       #plotting the count plot based on hour for 2012
plt.show()
```

```python
def plot_by_year(agg_attr, title):
    #extracting the required fields
    dd=train.copy()
    dd['year'] = train.datetime.dt.year  #extracting the year
    dd['month'] = train.datetime.dt.month   # " the month
    dd['hour'] = train.datetime.dt.hour  # " the hour
    by_year = dd.groupby([agg_attr,'year'])['count'].agg('sum').unstack()   #Group by year
    return by_year.plot(kind = 'bar', figsize = (15,5), width =0.9, title = title)   # Returning the figure grouped by year

plot_by_year('month', "Rent bikes per month in 2011 and 2012")       #plotting monthly bike rentals based on year
plot_by_year('hour', "Rent bikes per hour in 2011 and 2012")       #plotting hourly bike rentals based on year
plt.show()
```

```python
# method to plot a graph for count per hour box plot
def plot_hours(data, message = ''):
    dd = data.copy()
    dd['hour'] = data.datetime.dt.hour    #extracting the hour

    hours ={} 
    for hour in range(24):
        hours[hour] = dd[ dd.hour == hour ]['count'].values
    plt.figure(figsize = (20,10))
    plt.ylabel("count rent")
    plt.xlabel("hours")
    plt.title("count vs hours\n" + message)
    plt.boxplot([hours[hour] for hour in range(24)])

    axis = plt.gca()
    axis.set_ylim([1, 1100])
```

```python
plot_hours( train[train.datetime.dt.year == 2011], 'year 2011')   #box plot for hourly count for the mentioned year
plot_hours( train[train.datetime.dt.year == 2012], 'year 2011')   #box plot for hourly count for the mentioned year
plt.show()
```

```python
# method to convert catagorical data to numerical data
def catagorical_to_numeric(x):
    if 0 <= x<6:
        return 0
    elif 6 <= x < 13:
        return 1
    elif 13 <= x <19:
        return 2
    elif 19 <= x < 24:
        return 3
```

```python
train['hour'] = train['hour'].apply(catagorical_to_numeric)     #applying the above conversion logic to training data
train.head()
```
