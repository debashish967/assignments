# Variance

# Using NUMPY     ddof= delta degree of freedom

### If ddof = 0(population variance) , ddof = 1(sample variance)

```python
import numpy as np

data = [2,4,4,5,6,7,8,9]

sample_variance = np.var(data, ddof=1 )

pop_variance = np.var(data, ddof=0)

print ('the sample variance of the data is', sample_variance)
print ('the population variance of the data is', pop_variance)
```

# Using Pandas for Variance

```python
import pandas as pd
```

```python
data = [2,5,6,67,8,5,4,9,14,2,10]

series = pd.Series(data)

sample_variance = series.var()
print('the sample variance is', sample_variance)

pop_variance = series.var(ddof=0)
print('the population sample is', pop_variance)
                
```

# Using Scipy

```python
from scipy import stats
```

```python
data = [7,8,9,1,8,5,2,6,4,7]

sample_variance = stats.tvar(data)
print('the sample variance is', sample_variance)

popul_variance = stats.tvar(data, ddof= 0)
print('the population variance is', popul_variance)
```

# Standard Deviation

### using NUMPY

```python
import numpy as np

data = [60,65,70,75,80]

sample_std = np.std(data, ddof=1)
print ('the sample std of the data is', sample_std)

population_std = np.std(data, ddof = 0)
print('the population std of the data is', population_std)
```

### Using Pandas

```python
import pandas as pd

data = [2,5,6,67,8,5,4,9,14,2,10]
series = pd.Series(data)

sample_std = series.std()
print('the sample std of the data is', sample_std)

pop_std = series.std (ddof=0)
print ('the population std of the data is', pop_std)
```

# DATA REPRESENTATION

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and axes
fig, axs = plt.subplots (3, 2, figsize = (14, 12))
fig.suptitle('histograms with KDE of Various distributions', fontsize = 16)


# NormalDistribution (Gaussian Distribution)
data_normal = np.random.normal(loc = 0, scale = 1, size=1000)   #Mean=0, StdSev=1
sns.histplot(data_normal, kde = True, ax = axs[0, 0], color = 'blue', stat = 'density', bins = 30)
axs [0, 0].set_title('Normal Distribution')
axs[0, 0].grid(True)


# Skewed Distribution
data_right_skewed = np.random.exponential(scale=1, size = 1000)   # Right Skewed
data_left_skewed = -np.random.exponential(scale=1, size = 1000) # Left Skewed

# Right Skewed histogram
sns.histplot (data_right_skewed, kde=True, ax=axs[0, 1], color='red', stat = 'density', bins=30, label = 'Right skewed')
#Left skewed histrogram
sns.histplot(data_left_skewed, kde = True, ax=axs[0, 1], color = 'green', stat='density', bins= 30, label= 'Left skewed')
axs[0, 1].set_title('Skewed Distribution')
axs[0, 1].legend
axs[0, 1].grid(True)

# Uniform Distribution
data_uniform = np.random.uniform( low=0, high=1, size=1000) # Uniform distribution between 0 and 1
sns.histplot(data_uniform, kde=True, ax= axs[1, 0], color = 'Purple', stat = 'density', bins= 30)
axs[1, 0].set_title('Uniform Distribution')
axs[1, 0].grid(True)

# Bimodal Distribution
data_bimodal = np.concatenate([np.random.normal(loc=-2, scale=1, size=500),
                               np.random.normal(loc=2, scale=1, size=500)])
sns.histplot(data_bimodal, kde=True, ax=axs[1,1], color= 'orange', stat= 'density', bins= 30)
axs[1,1].set_title('Bimodal Distribution')
axs[1,1].grid(True)

# Multimodal Distribution
data_multimodal = np.concatenate([np.random.normal(loc=-6, scale=1.5, size= 500),
                                  np.random.normal(loc=0, scale=1, size=500),
                                  np.random.normal(loc= 8, scale=2, size=200)])
sns.histplot(data_multimodal, kde=True, ax=axs[2,0], color='brown' , stat= 'density', bins=30)
axs[2,0].set_title('Multimodal Distribution')
axs[2,0].grid(True)

# Hide empty subplot
axs[2, 1].axis('off')

# Adjust Layout
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

# T tset and P test

```python
import numpy as np
from scipy import stats

# Step 1: Create two random sample datasets
# Suppose these are test scores from two groups (e.g., group A and group B)
group_A = np.random.normal(75,10,30)  # mean 75, std 10, sample 30
group_B = np.random.normal(80,12,30)  # mean 80, std 12, sample 30

# Step 2: Perform an independent T-test
t_statistic, p_value = stats.ttest_ind(group_A, group_B)

# Step 3: Display the p-value and t-statistic
print (f"T-statistic: {t_statistic}")
print (f"p-value: {p_value}")

# Step 4: Interpretation based on p_value
alpha = 0.05  #Significance Level
if p_value <= alpha:
    print("Reject the null hypothesis (significant difference between groups)")
else: print("Fail to Reject the null hypothesis (No significant difference between groups)")
```
